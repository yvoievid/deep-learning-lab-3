```python

import matplotlib.pyplot as plt
from keras.layers import Conv2D, MaxPooling2D,  Flatten, Conv2D, Dense, Dropout
from keras.models import Sequential

from keras.utils import to_categorical
```

```python
plt.style.use('fivethirtyeight')
```

## Loading the dataset

```python
from keras.datasets import cifar10 
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print(type(x_train))
print(type(y_train))
print(type(x_test))
print(type(y_test))
```

```python
print('x_train shape:', x_train.shape)
print('y_train shape:', y_train.shape)
print('x_test shape:', x_test.shape)
print('y_test shape:', y_test.shape)
```

```python
print(x_train[:5])
```

```python
index = 0
img = plt.imshow(x_train[index])
```

```python
print('The image label is: ', y_train[index])
```

```python
classification = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'] 
classification[y_train[index][0]]
```

```python
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
```

```python
print(y_test_one_hot)
```

## Normalization

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
model = Sequential([
    Conv2D(filters=32, kernel_size=5, activation='relu', input_shape=(32,32,3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(filters=64, kernel_size=5, activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(1000, activation='relu'),
    Dropout(0.5),
    Dense(500, activation='relu'),
    Dropout(0.5),
    Dense(250, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train_one_hot, batch_size=256, epochs=10, validation_split=0.2)
```

## Testing

```python
model.evaluate(x_test, y_test_one_hot)
```

## Visualization of loss function

```python
import pandas as pd

hist_plot = pd.DataFrame(history.history)
hist_plot.loc[:,['loss', 'val_loss']].plot(title="Loss")
```

## Model Accuracy

```python
hist_plot = pd.DataFrame(history.history)
hist_plot.loc[:,['accuracy', 'val_accuracy']].plot(title="Accuracy", ylabel="Accuracy", xlabel="Epoch")
```
